```python
import numpy as np
import scipy.io

cn = 'case1/'
nt = 'networkD/'

# Load the .npy file
u_true = np.load(cn + nt + 'Results/u_true.npy')
u_pred = np.load(cn + nt + 'Results/u_pred.npy')

# Save the array as a .mat file
scipy.io.savemat(cn + nt + 'Results/pred.mat', {'u_true': u_true, 'u_pred': u_pred})
```
